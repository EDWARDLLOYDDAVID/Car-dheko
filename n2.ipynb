```python
import pandas as pd
import json
import re



# Initialize a list to hold all rows
rows = []

# Convert each string to a dictionary and extract the 'top' data
for car_str in df:
    # Replace single quotes with double quotes
    clean_str = car_str.replace("'", '"')
    # Remove commas from numeric values in the 'value' fields
    clean_str = re.sub(r'(\d),(\d)', r'\1\2', clean_str)
    # Replace any None with null (JSON requires null instead of Python's None)
    clean_str = clean_str.replace('None', 'null')
    
    try:
        # Load as JSON
        car_dict = json.loads(clean_str)
        rows.extend(car_dict['top'])
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
        print(f"Problematic string: {clean_str}")
        continue

# Convert the list of rows to a DataFrame
df = pd.DataFrame(rows)

# Display the DataFrame
print(df)
```

```python
import re

# Function to clean and convert price values to lakhs
def convert_price(price):
    price = price.strip()
    price = re.sub(r'[â‚¹,]', '', price)  # Remove currency symbols and commas
    if 'Lakh' in price:
        value = re.sub(r'[^\d.]', '', price)
        return float(value)
    elif 'Crore' in price:
        value = re.sub(r'[^\d.]', '', price)
        return float(value) * 100  # 1 Crore = 100 Lakhs
    elif re.match(r'^\d+(\.\d+)?$', price):  # If it's a pure number (thousands)
        return float(price) / 100000  # Convert thousands to lakhs (e.g., 80,000 to 0.8)
    else:
        return None

# Assuming df is your DataFrame and 'price' is the column with price values
df['price_in_lakhs'] = df['price'].apply(convert_price)

# Display the updated dataframe
print(df[['price', 'price_in_lakhs']].tail())
```

```python
from sklearn.preprocessing import LabelEncoder
from scipy.stats import zscore

# Identifying the categorical columns
categorical_columns = ['bt', 'Fuel Type', 'Insurance Validity', 'Transmission', 'City']

# Initializing the label encoder
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le.classes_

# Saving the label encoded values
labeled_values = {col: list(le) for col, le in label_encoders.items()}

# Calculating Z-scores for the numerical columns
df_zscore = df.copy()
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df_zscore[col] = zscore(df[col])

# Saving the labeled values and the dataset with Z-scores

df_zscore.to_csv('test.csv', index=False)

labeled_values, df_zscore
```
